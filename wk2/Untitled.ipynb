```r
notes <- read.csv("../data/notas.csv" )  
notes
```

```r
notes['Condicion'] <- notes$notas

i <- 1

nrows = dim(notes)[1]

while (i <= nrows){
    
     if (notes$notas[i] >= 11){
        notes$Condicion[i] = 'Aprobo'
        }
    else if (notes$notas[i] < 11){
        notes$Condicion[i] = 'No aprobo'
    }
    i <- i + 1 
}
notes
```

```r
nrows <- dim(notes)[1]
i <- 1

while (i < nrows){
    
    if (notes$carrera[i] == 'a001'){
      notes$nombre_carrera[i] = 'Economía'
}
    else if (notes$carrera[i] == 'a002'){ 
             notes$nombre_carrera[i] = 'Química'
    } 
    else if (notes$carrera[i] == 'a003'){
             notes$nombre_carrera[i] = 'Filosofía'
    } 
    else {
        notes$nombre_carrera[i] = '-'
    }
    i <- i + 1
}

notes
```
