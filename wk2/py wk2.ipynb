```python
import pandas as pd
notes = pd.read_csv( r"..\data\notas.csv" )
notes
```

```python
notes['Condicion'] = notes.notas

i = 1

nrows = notes.notas.count()

while i <= nrows:
    
     if notes['notas'][i] >= 11:
        notes.Condicion[i] = 'Aprobo'
    
    elif notes['notas'][i] < 11:
        notes.Condicion[i] = 'No aprobo'
    i <- i + 1 
notes
```
